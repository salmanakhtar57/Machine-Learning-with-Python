```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_accidents = sns.load_dataset(name ='car_crashes', index_col='abbrev')[['alcohol', 'total']]
df_accidents.rename({'total': 'accidents'}, axis=1, inplace=True)
df_accidents 
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model_lr = LinearRegression()
```

```python
model_lr.fit()
```

```python
target = df_accidents['accidents']
features = df_accidents[['alcohol']]
```

```python
model_lr.fit(X=features, y=target)
```

```python
features
```

```python
target
```

```python
model_lr.predict(X=features)
```

```python
df_accidents['pred_lr'] = model_lr.predict(X=features)
df_accidents
```

```python
sns.scatterplot(x='alcohol', y='accidents', data=df_accidents)
sns.scatterplot(x='alcohol', y='pred_lr', data=df_accidents);
```

```python
sns.scatterplot(x='alcohol', y='accidents', data=df_accidents)
sns.scatterplot(x='alcohol', y='pred_lr', data=df_accidents);
sns.lineplot(x='alcohol', y='pred_lr', data=df_accidents, color='orange');
```

```python
#To measure the quality of the model, we use the .score() function
#to correctly calculate the difference between the model's predictions and reality.

model_lr.score(X=features, y=target)
```

Explain the Score

```python
df_accidents['accidents'] - df_accidents['pred_lr']
```

```python
df_accidents['residuals'] = df_accidents['accidents'] - df_accidents['pred_lr']
df_accidents
```

```python
df_accidents.residuals.sum()
```

```python
df_accidents.residuals.sum().round(2)
```

```python
df_accidents['residuals^2'] = df_accidents.residuals**2
df_accidents
```

```python
df_accidents['residuals^2'].sum()
```

```python
RSS = df_accidents['residuals^2'].sum()
RSS
```

```python
df_accidents.accidents
```

```python
df_accidents.accidents.mean()
```

```python
df_accidents['real_residuals'] = df_accidents.accidents - df_accidents.accidents.mean()
df_accidents
```

```python
df_accidents['real_residuals^2'] = df_accidents.real_residuals**2
```

```python
df_accidents['real_residuals^2'].sum()
```

```python
TSS = df_accidents['real_residuals^2'].sum()
```

```python
RSS/TSS
```

```python
1 - RSS/TSS
```

RandomForestRegressor() in Python

```python
from sklearn.ensemble import RandomForestRegressor

model_rf = RandomForestRegressor()
model_rf.fit(X=features, y=target)
```

```python
model_rf.predict(X=features)
```

```python
df_accidents['pred_rf'] = model_rf.predict(X=features)
```

```python
model_rf.score(X=features, y=target)
```

```python
dic_scores = {}
dic_scores['lr'] = model_lr.score(X=features, y=target)
dic_scores['rf'] = model_rf.score(X=features, y=target)
```

SVR() in Python

```python
from sklearn.svm import SVR

model_sv = SVR()
model_sv.fit(X=features, y=target)
```

```python
SVR()
```

```python
model_sv.predict(X=features)
```

```python
df_accidents['pred_sv'] = model_sv.predict(X=features)
```

```python
model_sv.score(X=features, y=target)
```

```python
dic_scores['sv'] = model_sv.score(X=features, y=target)
```

```python
pd.Series(dic_scores).sort_values(ascending=False)
```
